```python
gdfmerged = gdf.merge(df, on='ntaname', how='inner')
gdfmerged = gdfmerged[['nta2020','geometry','Population','boroname']]
gdfmerged.head(3)
```

```python
f, ax = plt.subplots(1,1,figsize=(5,5))
gdfmerged.plot(ax=ax, column="Population", cmap='Reds')
ax.axis('off')
```

```python
gdf_manhattan = gdfmerged[gdfmerged.boroname=="Manhattan"]
```

```python
gdf_manhattan.head(3)
```

```python
gdf_manhattan_local = gdf_manhattan.to_crs(2263)
gdfmerged = gdfmerged.to_crs(2263)
```

```python
gdf_manhattan_local.plot()
```

```python
distances = [100, 1000, 5000, 10000, 25000, 100000]
for d in distances:
    gdf_manhattan_local.buffer(d).plot()
```

```python
gdfmerged.plot()
```

```python
gdfmerged_points = gdfmerged.copy()
gdfmerged_points['geometry'] = [g.centroid for g in list(gdfmerged_points.geometry)]
gdfmerged_points.head(3)
```

```python
gdfmerged_points.plot()
```

```python
distance = 100000
gdf_manhattan_local_buffered = gdf_manhattan_local.copy()
gdf_manhattan_local_buffered.plot()
```

```python
gdf_manhattan_local_buffered['geometry'] = gdf_manhattan_local_buffered['geometry'].buffer(distance)
gdf_manhattan_local_buffered.plot()
```

```python
gdfmerged_points.head(3)
```

```python
print(len(gdf_manhattan_local_buffered))
print(len(gdfmerged_points))
gdf_manhattan_local_buffered.head(3)
```

```python
tdf = pd.read_csv('2015_Street_Tree_Census_-_Tree_Data_20251111.csv')
```

```python
len(tdf)
```

```python
tdf.head(3)
```

```python
from collections import Counter
Counter(tdf.status).most_common()
```

```python
Counter(tdf.health).most_common()
```

```python
from shapely.geometry import Point
```

```python
geometry = [Point(xy) for xy in zip(tdf['longitude'], tdf['latitude'])]
geometry[0]
```

```python
gdf_trees = gpd.GeoDataFrame(tdf, geometry=geometry)[['tree_id', 'stump_diam', 'status', 'health','geometry']]
gdf_trees.crs = 4326
```

```python
gdf_trees.head(3)
```

```python
f, ax = plt.subplots(1,1,figsize=(10,10))

gdf_trees.plot(ax=ax, column='status', legend=True, markersize=2, alpha=0.4)
ax.axis('off')
```
