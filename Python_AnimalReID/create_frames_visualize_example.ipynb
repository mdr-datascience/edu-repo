```python
# Import packages
import cv2
import os
import matplotlib.pyplot as plt

# Define the path to the video file
video_path = 'videos\\videos\\Koi_5652_952_540.mp4'

# Define the directory to save the frames
output_dir = 'Koi_frames'
os.makedirs(output_dir, exist_ok=True)

# Open the video file
cap = cv2.VideoCapture(video_path)

# Check if the video file opened successfully
if not cap.isOpened():
    print(f"Error: Unable to open video file {video_path}")
else:
    frame_count = 1
    while True:
        # Read a frame from the video
        ret, frame = cap.read()
        
        # If the frame was not read successfully, break the loop
        if not ret:
            break
        
        # Define the path to save the current frame
        frame_path = os.path.join(output_dir, f'frame_{frame_count:04d}.jpg')
        
        # Save the frame as an image file
        cv2.imwrite(frame_path, frame)
        
        # Increment the frame count
        frame_count += 1

    # Release the video capture object
    cap.release()
    print(f"Frames saved to {output_dir}")


# Read labels (more on the .csv files called frames)
x, y, width, height = 556, 267, 324, 213
max_w, max_h, label = 952, 540, 6
frame = '48'

# Load the image
image_path = output_dir + '\\frame_00' + frame + '.jpg'
image = cv2.imread(image_path)

# Check if the image is loaded correctly
if image is None:
    print(f"Error: Unable to load image at {image_path}")
else:
    # Draw the bounding box
    top_left = (x, y)
    bottom_right = (x + width, y + height)
    color = (255, 0, 0)  # Red color for the bounding box
    thickness = 2

    cv2.rectangle(image, top_left, bottom_right, color, thickness)

    # Display the image with the bounding box
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"Label: {label}")
    plt.axis('off')
    plt.show()
```
