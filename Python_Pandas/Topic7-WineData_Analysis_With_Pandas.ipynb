# Wine Dataset Analysis with Pandas

We will see the different ways to select subsets of data in Pandas using [], .loc, .iloc, etc. We will be
using the wine quality dataset hosted on the UCI website. This data record 11 chemical properties
(such as the concentrations of sugar, citric acid, alcohol, pH etc.) of thousands of red and white
wines from northern Portugal, as well as the quality of the wines, recorded on a scale from 1 to 10.
We will only look at the data for red wine.

```python
import pandas as pd

# Load the data directly from the internet.
wine_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
```

```python
# Print the first few lines of the dataframe
wine_df.head()
```

## Different ways to select columns

```python
# Print the sequence of columns as desired
wine_df.columns = ['fixed_acidity', 'volatile_acidity', 'citric_acid',
'residual_sugar', 'chlorides', 'free_sulfur_dioxide',
'total_sulfur_dioxide','density','pH',
'sulphates', 'alcohol', 'quality' ]
```

```python
# Find the first few entries of the column - fixed_acidity!
wine_df['fixed_acidity'].head()
```

```python
# Another way of accessing the same column
wine_df.fixed_acidity.head()
```

## Selecting multiple columns

```python
# Select a set of columns - Just mention their names!
wine_four = wine_df[['fixed_acidity', 'volatile_acidity','citric_acid','residual_sugar']]
wine_four.head()
```

```python
# Alternately, use cols as list variable sa well
cols = ['fixed_acidity', 'volatile_acidity','citric_acid', 'residual_sugar']
wine_list_four = wine_four[cols]
```

```python
# Print the first few lines
wine_list_four.head()
```

## Selecting columns using “select_dtypes” and “filter” methods

```python
# Find the count of unique data types in the dataframe.
wine_df.dtypes.value_counts()
```

```python
# Find the float columns in the dataset
wine_df.select_dtypes(include = ['float']).head()
```

```python
# find all the numerical features in the dataset!
wine_df.select_dtypes(include=['number']).head()
```

```python
# Find all the columns with "acid" in its feature name.
wine_df.filter(like='acid').head()
```

```python
# Change the name of couple of columns
wine_df.rename(columns={'pH':'pH_5', 'quality': 'quality_6' }, inplace=True)
wine_df.filter(regex='\d').head()
```

```python
# Did it change the original dataframe.
# What did the inplace parameter do?
wine_df.head()
```

```python
# Filter the columns with the "fixed_acidity" and "volatile_acidity" in its name!
wine_df.filter(items=['fixed_acidity', 'volatile_acidity']).head()
```

```python
# Find the top few lines of the dataframe!
wine_df.head()
```

## Changing the order of your columns

It is very easy to change the column order of the dataframes.

```python
# Find the feature names of the dataset.
wine_df.columns
```

```python
# Change the order of the columns by splitting into groups.
group_1 = ['pH_5','sulphates','alcohol', 'quality_6']
group_2 =['chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density']
group_3 = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar']
```

```python
# Combine these groups together.
new_cols = group_1+group_2+group_3
```

```python
# Print he column names
wine_df.columns
```

```python
# If we use the sets datatype for comparison, we get same sets
# This ensures all features are covered in the final set.
set(wine_df.columns) == set(new_cols)
```

```python
# As such the order is different. Check that as comparing lists.
list(wine_df.columns) == list(new_cols)
```

```python
# Create a new dataframe with the new column sequence.
wine_df_2 = wine_df[new_cols]
```

```python
# Print few lines and see if the order of columns changed?
wine_df_2.head()
```

## Selecting rows using .iloc and loc

Rows can be extracted separatly using index functionality.

```python
# Find the number of unique entries across each feature.
wine_df.nunique()
```

```python
# Drop some values based on the unique density feature.
wine_df = wine_df.drop_duplicates(subset='density')
```

```python
# How many entries exists?
wine_df.shape #436 unique values in the density column
```

```python
#Since density has only unique values, it can be used as index.
wine_df.set_index('density',inplace=True)
```

```python
# Check if the index changed!
wine_df.head()
```

```python
# Access the rows using iloc function.
wine_df.iloc[2]
```

```python
# Select the particular wine with density 0.9968!
wine_df.loc[0.9968]
```

```python
# Select only a few selected rows of data
wine_df.iloc[[1, 4, 7]]
```

```python
# If you find the index names, iloc may not be needed.
rows = [0.9968, 0.9964, 0.9943]
wine_df.loc[rows]
```

```python
# iloc extract a slice of data
wine_df.iloc[1:4]
```

```python
# Could use variables instead.
first = 0.9970
last = 0.9959
wine_df.loc[first:last]
```

## Selecting rows and columns simultaneously

```python
# Again see top entries
wine_df.head()
```

```python
# Use iloc to select some columns
wine_df.iloc[:, [3,4,6]].head()
```

```python
# Select the features by name even!
wine_df.loc[:,['residual_sugar','chlorides','total_sulfur_dioxide']].head()
```

## Selecting disjointed rows and columns

```python
# Select a combination of rows and columns
wine_df.iloc[[10,14], [7, 9]]
```

```python
# Even as a variable.
rows = [0.9969, 0.9962]
columns = ['pH_5', 'alcohol']
wine_df.loc[rows,columns]
```

```python
# Print a density and the corresponding chloride value.
wine_df.loc[0.9970, 'chlorides']
```

```python
# Even select a particular combination
wine_df.iloc[2, 4]
```

```python
# Check the top 10 entries!
wine_df.head(10)
```

```python
# Even as a slice as index
wine_df.iloc[8:2:-2, 2]
```

```python
# Use the density indices as values..
first = 0.9974
second = 0.9964
wine_df.loc[first:second:-2, 'citric_acid']
```

```python
# Select first five rows
wine_df.iloc[:5]
```

```python
# Equivalently...
wine_df.iloc[:5, :]
```

## Selecting rows and columns using get_loc and index methods

```python
# Find the start and stop in that counts.
col_start = wine_df.columns.get_loc('volatile_acidity')
col_end = wine_df.columns.get_loc('volatile_acidity')+2
col_start, col_end
```

```python
# Use these new start and end as indices
wine_df.iloc[:4, col_start:col_end]
```

```python
# select the wines with the range of indicies
row_start = wine_df.index[4]
row_end = wine_df.index[7]
wine_df.loc[row_start:row_end, 'volatile_acidity' : 'chlorides']
```

```python
# Find a specific combination
density_val = 0.9980
wine_df.loc[density_val, 'citric_acid']
```
