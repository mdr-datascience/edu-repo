# Seaborn Visualisations

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

## 1. Data

The beauty of Seaborn is it comes with built-in data sets!

```python
# Load the dataset
penguins = sns.load_dataset("penguins")
```

## 2. Distribution Plots

Following are different plots that allow us to visualize the distribution of a data set. These plots
are:
- distplot
- jointplot
- pairplot
- rugplot
- kdeplot

For example:

### 2.1 distplot

The distplot shows the distribution of a univariate set of observations.

```python
# Create a distribution plot
sns.displot(data=penguins, x="flipper_length_mm", kind="hist", kde=True)

plt.show()
```

### 2.2 jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of
what 'kind' parameter to compare with: * “scatter” * “reg” * “resid” * “kde” * “hex”.


```python
# Create a joint plot
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
plt.show()
```

```python
# Change kind to hex
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind='hex')
plt.show()
```

```python
# Change kind to reg
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind='reg')
plt.show()
```

### 2.4 pairplot
pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns)
and supports a color hue argument (for categorical columns).

```python
# Plot the penguin data pairwise, use the variable sex as color hue
sns.pairplot(penguins,hue='sex',palette='coolwarm')
plt.show()
```

## 3 Categorical Data 

Now let’s discuss using seaborn to plot categorical data! There are a few main plot types for this:
- factorplot
- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

Let’s go through examples of each!


```python
# Load tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

### 3.1 barplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.show()
```

```python
# You can change the estimator object to your own function, that converts a vector to a scalar:
import numpy as np

sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
plt.show()
```

### 3.2 countplot

```python
sns.countplot(x='sex',data=tips)
plt.show()
```

### 3.3 boxplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons
between variables or across levels of a categorical variable. The box shows the quartiles of the
dataset while the whiskers extend to show the rest of the distribution, except for points that are
determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
plt.show()
```

## 4. Matrix Plot

Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate
clusters within the data.

Let’s begin by exploring seaborn’s heatmap and clutermap:

```python
# Load two datasets

flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

### 4.1 Heatmap

In order for a heatmap to work properly, your data should already be in a matrix form, the
sns.heatmap function basically just colors it in for you. For example:

```python
tips.head()
```

```python
# Calculate correlation matrix for numerical variables
correlation = tips[['total_bill', 'tip', 'size']].corr()
```

```python
# Plot correlation matrix
sns.heatmap(correlation)
plt.show()
```

```python
# Try a different colour palette
sns.heatmap(correlation,cmap='coolwarm',annot=True)
plt.show()
```

```python
# Try the other dataset
pv_flights = flights.pivot_table(values='passengers',index='month',columns='year',observed=True)
pv_flights
```

```python
sns.heatmap(pv_flights)
plt.show()
```

### 4.2 clustermap

The clustermap uses hierarchal clustering to produce a clustered version of the heatmap. For
example:

```python
sns.clustermap(pv_flights)
plt.show()
```

Notice now how the years and months are no longer in order, instead they are grouped by
similarity in value (passenger count). That means we can begin to infer things from this plot, such
as August and July being similar (makes sense, since they are both summer travel months).
